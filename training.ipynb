```python
import os
import random
import shutil

# Set paths to your images and labels folders
images_folder = "./images"
labels_folder = "./labels"

# Set paths for the output folders
output_train_images_folder = "./train/images"
output_val_images_folder = "./val/images"
output_train_labels_folder = "./train/labels"
output_val_labels_folder = "./val/labels"

# Ensure output directories exist, create them if not
os.makedirs(output_train_images_folder, exist_ok=True)
os.makedirs(output_val_images_folder, exist_ok=True)
os.makedirs(output_train_labels_folder, exist_ok=True)
os.makedirs(output_val_labels_folder, exist_ok=True)

# Get lists of image and label filenames
image_filenames = os.listdir(images_folder)
label_filenames = os.listdir(labels_folder)

# Shuffle the filenames to randomize the split
random.shuffle(image_filenames)

# Calculate the number of images for training and validation (80/20 split)
num_train_images = int(0.8 * len(image_filenames))
num_val_images = len(image_filenames) - num_train_images

# Split image filenames into training and validation sets
train_image_filenames = image_filenames[:num_train_images]
val_image_filenames = image_filenames[num_train_images:]

for image_filename in train_image_filenames:
    label_filename = image_filename.replace(".jpg", ".txt")  # Assuming label files have the same name but with a different extension
    shutil.copy(os.path.join(images_folder, image_filename), output_train_images_folder)
    shutil.copy(os.path.join(labels_folder, label_filename), output_train_labels_folder)

for image_filename in val_image_filenames:
    label_filename = image_filename.replace(".jpg", ".txt")  # Assuming label files have the same name but with a different extension
    shutil.copy(os.path.join(images_folder, image_filename), output_val_images_folder)
    shutil.copy(os.path.join(labels_folder, label_filename), output_val_labels_folder)

```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Function to get image files from a folder
def get_image_files(folder_path):
    image_files = []
    for file in os.listdir(folder_path):
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            image_files.append(os.path.join(folder_path, file))
    return image_files

# Path to the folder containing images
folder_path = "./newdata/annotatedimages"

# Get image paths from the folder
image_paths = get_image_files(folder_path)

# Plotting only the first 9 images
plt.figure(figsize=(12, 18))
columns = 3
rows = 3
for i in range(min(9, len(image_paths))):  # Plot only the first 9 images
    img = mpimg.imread(image_paths[i])
    plt.subplot(rows, columns, i + 1)
    plt.imshow(img)
    plt.axis('off')

plt.show()
```
